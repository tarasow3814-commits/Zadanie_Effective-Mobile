```python
Задание 1
Задание 1: Фермер
На ферме содержатся шесть разных видов животных, и каждый раз, когда фермер заходит в сарай, он видит одно случайное животное. За день фермер заходит в сарай 6 раз.
Каково математическое ожидание количества разных видов животных, которые фермер увидит за день?
Ответ округлить до сотых
```

```python
import random

def simulate_day(num_animals=6, num_visits=6, num_simulations=100000):
    """
    Симуляция множества дней для оценки математического ожидания.

    Параметры:
    - num_animals: количество видов животных (6)
    - num_visits: количество посещений за день (6)
    - num_simulations: количество симуляций для усреднения

    Возвращает:
    - Оценку математического ожидания количества увиденных видов
    """
    total_unique_seen = 0

    for _ in range(num_simulations):
        seen_animals = set()

        for _ in range(num_visits):
            # Случайный выбор животного (от 0 до 5)
            animal = random.randint(0, num_animals - 1)
            seen_animals.add(animal)

        # Количество уникальных животных за этот день
        total_unique_seen += len(seen_animals)

        expected_value = total_unique_seen / num_simulations
    return expected_value

result = simulate_day(num_simulations=100000)
print(f"Оценённое математическое ожидание: {result:.2f}")
```

```python
Задание 2: Кулинарное соревнование
В конкурсе участвуют 80 шеф-поваров с уникальными уровнями мастерства. В первом этапе судьи случайным образом распределяют их по парам (в любом состязании двух шефов выигрывает тот, у кого выше уровень мастерства). На втором этапе шефы снова случайно образуют пары для финального раунда (пары могут повториться). Победную награду получают те, кто выиграл в обоих этапах.
Каково математическое ожидание числа победителей?
Ответ округлить до десятых
```

```python
# Количество шеф-поваров
n = 80

# Знаменатель (количество возможных соперников)
denominator = n - 1  # 79

# Вычисляем сумму квадратов чисел от 0 до 79 (т.е. (k-1)^2)
sum_of_squares = sum(m ** 2 for m in range(denominator + 1))

# Математическое ожидание
expected_winners = sum_of_squares / (denominator ** 2)

# Округление до десятых
result = round(expected_winners, 1)

print(f"Математическое ожидание числа победителей: {result}")
```

```python
Задание 3: Одинокая дорога
На пустынном шоссе вероятность появления автомобиля за 30-минутный период составляет 0.95.
Какова вероятность его появления за 10 минут? А за 27 минут?
Ответ дать в процентах, округлив до десятых через точку с запятой
```

```python
import pandas as pd
import numpy as np

# Исходные данные
P_30 = 0.95  # вероятность появления авто за 30 минут
no_car_30 = 1 - P_30  # вероятность отсутствия авто за 30 минут (0.05)

# Времена, для которых нужно рассчитать вероятности
times = [10, 27]

# Функция для расчёта вероятности появления авто за t минут
def calculate_probability(t):
    # Формула: P(t) = 1 - (вероятность отсутствия за 30 мин)^(t/30)
    P_no_car_t = no_car_30 ** (t / 30)
    return 1 - P_no_car_t

# Создаём список для хранения результатов
results = []

# Рассчитываем вероятности для каждого времени
for t in times:
    P_t = calculate_probability(t)
    P_percent = round(P_t * 100, 1)  # в процентах, округлено до десятых
    results.append({'Время (мин)': t, 'Вероятность (%)': P_percent})

# Создаём DataFrame
df_results = pd.DataFrame(results)

# Выводим таблицу
print("Вероятности появления автомобиля на шоссе:")
print(df_results.to_string(index=False))

# Форматируем итоговый ответ в требуемом виде: 63.2;93.3
final_result = ';'.join(map(str, df_results['Вероятность (%)'].tolist()))
print(f"\nОтвет: {final_result}")
```

```python
Блок 2. Phyton
Задание 1. Реализовать функцию (или тело функции), 
которая проверяет на изоморфность два слова.
Пояснение: строки s и t называются изоморфными, если все вхождения каждого символа 
строки s можно последовательно заменить другим символом и получить строку t.
Порядок символов при этом должен сохраняться, а замена — быть уникальной.
Так, два разных символа строки s нельзя заменить одним и тем же символом из строки t, 
а вот одинаковые символы в строке s должны заменяться одним и тем же символом.
```

```python
def is_isomorphic(s: str, t: str) -> bool:
    # Если длины строк не совпадают, они не могут быть изоморфными
    if len(s) != len(t):
        return False

    # Словари для хранения соответствий символов
    s_to_t = {}  # Соответствие из s в t
    t_to_s = {}  # Соответствие из t в s

    for char_s, char_t in zip(s, t):
        # Проверяем соответствие s -> t
        if char_s in s_to_t:
            if s_to_t[char_s] != char_t:
                return False
        else:
            s_to_t[char_s] = char_t

        # Проверяем соответствие t -> s
        if char_t in t_to_s:
            if t_to_s[char_t] != char_s:
                return False
        else:
            t_to_s[char_t] = char_s

    return True

s = 'paper'
t = 'title'
print(is_isomorphic(s, t))  
```

```python
Задание 2: Натуральная последовательность
Реализовать функцию (или тело функции), которая находит единственное отсутствующее число
из последовательности натуральных чисел 1,2,…,n
```

```python
def missing_number(nums):
    n = len(nums) + 1  # Общее количество чисел в полной последовательности
    expected_sum = n * (n + 1) // 2  # Сумма 1+2+...+n
    actual_sum = sum(nums)  # Сумма имеющихся чисел
    return expected_sum - actual_sum

# Пример использования
nums = [1, 2, 3, 4, 5, 6, 8, 9, 10, 11]
print(missing_number(nums)) 
```

```python
Задание 3: Факторизация
Реализовать функцию (или тело функции), которая при введении натурального числа n 
разбивает его на простые множители (представить его в виде простых чисел).
```

```python
def factorize(n):
    if n < 2:
        raise ValueError("Число должно быть >= 2")
    
    factors = []
    divisor = 2
    
    while divisor * divisor <= n:
        while n % divisor == 0:
            factors.append(divisor)
            n //= divisor
        divisor += 1
    
    # Если n > 1, значит, оставшийся n — простое число
    if n > 1:
        factors.append(n)
    
    return factors

# Примеры использования
print(factorize(12))   # [2, 2, 3]
print(factorize(17))   # [17]
print(factorize(60))   # [2, 2, 3, 5]
print(factorize(97))   # [97]
```
