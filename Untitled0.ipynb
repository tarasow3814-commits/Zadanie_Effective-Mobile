```python
import matplotlib.pyplot as plt
import numpy as np

# Данные
current_value = -12.59  # текущее значение (%)
target_value = 12.64    # целевое значение (%)
min_limit = -15
max_limit = 15

# Преобразование значения в угол (от -90 до +90 градусов)
# -15% -> -90° (лево, 9 часов)
# 0%   -> 0°   (верх, 12 часов)
# +15% -> +90° (право, 3 часа)
def value_to_angle(value):
    return (value - min_limit) / (max_limit - min_limit) * 180 - 90

angle_current = value_to_angle(current_value)
angle_target = value_to_angle(target_value)
angle_zero = value_to_angle(0)

# Создаём полярный график
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(projection='polar')

# Устанавливаем: 0° будет вверху (12 часов)
ax.set_theta_zero_location('N')  # север вверху
ax.set_theta_direction(-1)       # по часовой стрелке

# Углы от -90 до +90 (полукруг вниз)
ax.set_thetamin(-90)
ax.set_thetamax(90)

# Рисуем фоновую шкалу (серая) - без чёрных линий
theta_bg = np.radians(np.linspace(-90, 90, 100))
r_bg = [0.8] * len(theta_bg)
ax.plot(theta_bg, r_bg, color='lightgray', linewidth=40, alpha=0.3)

# Рисуем красный сектор от 0% до текущего значения
if current_value < 0:
    theta_red_deg = np.linspace(angle_current, angle_zero, 100)
    theta_red_rad = np.radians(theta_red_deg)

    for i in range(len(theta_red_rad) - 1):
        ax.fill_between([theta_red_rad[i], theta_red_rad[i+1]],
                        0.6, 1.0,
                        color='red', alpha=0.8, edgecolor='none')

# Рисуем зелёную полоску для целевого значения
ang_targ_rad = np.radians(angle_target)
ax.plot([ang_targ_rad, ang_targ_rad], [0.6, 1.0], color='green', linewidth=4,
        alpha=0.9, zorder=15)

# Только внешний контур шкалы (без внутренних чёрных линий)
theta_outline = np.radians(np.linspace(-90, 90, 100))
r_outline = [0.8] * len(theta_outline)
ax.plot(theta_outline, r_outline, color='black', linewidth=2)

# Добавляем ТОЛЬКО три деления: -15%, 0%, +15% (только риски, без чёрных линий по шкале)
tick_values = [-15, 0, 15]
tick_angles = [value_to_angle(v) for v in tick_values]

for val, ang_deg in zip(tick_values, tick_angles):
    ang_rad = np.radians(ang_deg)
    # Только риска (короткая чёрточка), без чёрной линии по шкале
    ax.plot([ang_rad, ang_rad], [0.72, 0.88], color='black', linewidth=2.5)
    # Подпись значения
    r_text = 0.96
    ax.text(ang_rad, r_text, f'{val}%', ha='center', va='center',
            fontsize=11, fontweight='bold')

# Отмечаем текущее значение (красный маркер)
ang_curr_rad = np.radians(angle_current)
ax.scatter(ang_curr_rad, 0.9, color='red', s=200, marker='v',
           zorder=10, edgecolor='darkred', linewidth=2)

# Крупное значение в центре - КРАСНЫМ
ax.text(0, 0, f'{current_value}%', ha='center', va='center',
        fontsize=28, fontweight='bold', color='red', zorder=20)

# Подпись для целевого значения
ax.text(ang_targ_rad, 0.52, f'Цель: {target_value}%', ha='center', va='top',
        fontsize=9, color='green', fontweight='bold')

# Отметка 0 (вверху)
ax.scatter(np.radians(angle_zero), 0.9, color='black', s=150,
           marker='o', zorder=10, edgecolor='black', linewidth=2.5)
ax.scatter(np.radians(angle_zero), 0.9, color='white', s=50,
           marker='o', zorder=11, edgecolor='black', linewidth=1)

# Оформление
ax.set_ylim(0, 1.1)
ax.set_yticklabels([])
ax.set_xticklabels([])
ax.spines['polar'].set_visible(False)
ax.grid(False)

# Заголовок
plt.suptitle('Датчик значения', fontsize=16, fontweight='bold', y=0.98)

plt.tight_layout()
plt.show()
```
